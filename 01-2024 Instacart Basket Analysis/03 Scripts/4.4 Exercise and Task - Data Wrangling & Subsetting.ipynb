# 10. Subsetting

```python
df_prods['department_id']==19
```

```python
#Enclosing df_prods['department_id']==19 inside df_prods[] tells Python to map the true/false procedure you witnessed -
# in the previous output onto the columns within the df_prods dataframe—the result being a list of only those values - 
# within df_prods that are true (and, subsequently, only those values within df_prods that have a "department_id" of 19)

df_prods[df_prods['department_id']==19] 
```

```python
# creating a subset that filters only the products of the 'scacks' department within 'df_prods' dataframe

# Within the brackets on the right-hand side, tell Python to look into the df_prods dataframe and find a column called "department_id.
# And, within that "department_id" column, it should look for a value of 19.

df_snacks =  df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# a second way to achieve these sames results is using the 'loc' function, to look for a specific column name

df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
df_snacks_2.head()
```

```python
# there is also a third command to get to the resulst above 
## In this second version, rather than using an equals sign, you’re telling the loc function to look - 
## into a list: isin([19]) This is handy because you could use more than one value here if you wanted - 
## to check multiple "department_ids" at the same time (isin([17,18,19])).

df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```

```python
df_snacks_3.head()
```
