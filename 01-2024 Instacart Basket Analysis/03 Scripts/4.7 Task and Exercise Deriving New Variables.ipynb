# 4. Define a function - User-defined function

### Create a new function using the 'if' statement

```python
# Define a new function to assign a price range (new variable) for the products

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the new function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check counts

df['price_range'].value_counts()
```

# 5. Using 'loc' function

### Instead of creating a new function, use the 'loc' function to achieve the same results.

### Sometimes, user-defined functions are the only way to accomplish something you need to do in Python. However, this isn’t always the case. As you saw in the warning message above, Python has suggested using the predefined function loc() to accomplish the same thing.

```python
# execute the 'loc' function with a conditional logic of an if-statement to a function 
## without explicitly creating an if-else construct

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# check the counts of the variable extracted from the function above

df['price_range_loc'].value_counts(dropna = False)
```

```python
# execute the 'loc' function (now on the entire data set)

ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High_range product'
```

### Find the busiest periods of time.
### Label them “Most orders,” “Average orders,” and “Fewest orders.” 
### Create a new column containing these labels called “busiest_period_of_day.”

```python
# check the counts values to find the 'busiest period od day' for orders
ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Use a 'loop' function to compare the values with what are the busiest and period of day for orders,
## and assign it the corresponding string value.

result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12:
    result.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

###  - The best and most accurate approach to achieving this result is to establish some thresholds by applying the percentile method to the counts of values recorded for each hour of the day.

### - Then, execute the 'loc' function with conditional logic, like an if-statement, containing the  thresholds mentioned above.

```python
# Create a new column that contains only the hour component of the timestamp

order_counts_per_hour = { 10: 2761760, 11: 2736140, 14: 2689136, 15: 2662144, 13: 2660954,
    12: 2618532, 16: 2535202, 9: 2454203, 17: 2087654, 8: 1718118,
    18: 1636502, 19: 1258305, 20: 976156, 7: 891054, 21: 795637,
    22: 634225, 23: 402316, 6: 290493, 0: 218769, 1: 115700,
    5: 87961, 2: 69375, 4: 53242, 3: 51281}

# Determine thresholds

most_orders_threshold = pd.Series(order_counts_per_hour).quantile(0.8)
fewest_orders_threshold = pd.Series(order_counts_per_hour).quantile(0.2)

# Initialize the new column with 'Average orders'
ords_prods_merge['busiest_period_of_the_day'] = 'Average orders'

# Classify each hour of day
ords_prods_merge.loc[ords_prods_merge['order_hour_of_day'].map(order_counts_per_hour) >= most_orders_threshold, 'busiest_period_of_the_day'] = 'Most orders'
ords_prods_merge.loc[ords_prods_merge['order_hour_of_day'].map(order_counts_per_hour) <= fewest_orders_threshold, 'busiest_period_of_the_day'] = 'Fewest orders'
```
