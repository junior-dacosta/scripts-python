## 2.1 Renaming Columns

```python
# Rename columns to clarity

df_customers.rename(columns={'First Name': 'first name',
                              'Surnam': 'last name',
                              'Gender': 'gender',
                              'STATE': 'state',
                              'Age': 'age',
                              'date_joined': 'customer_since'}, inplace=True)
```

```python
# Check values

df_customers.describe()
```

## 2.3 Changing data type 

```python
# Change data type from 'user_id' column to 'string',
## 'age' and 'n_dependants' to int16


df_customers['user_id'] = df_customers['user_id'].astype('str')

df_customers['age'] = df_customers['age'].astype('int16')

df_customers['n_dependants'] = df_customers['n_dependants'].astype('int16')
```

```python
# Check values again

df_customers.describe()
```

## 2.3 Mixed Data Type

```python
# Check if there are mixed data types

for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
     print (col)
```

```python
df_customers['first name'].dtype
```

```python
# Investigate the mixed data type in 'first name' column

for value in df_customers['first name']:
    if not isinstance(value, str):
        print(value)  # This will print out non-string values
```

# 3. Missing Values

```python
## Replace NaN in 'first' column with 'Unknown'
# Since the last name column contains data and the others columns too

df_customers['first name'].fillna('Unknown', inplace=True)
```
