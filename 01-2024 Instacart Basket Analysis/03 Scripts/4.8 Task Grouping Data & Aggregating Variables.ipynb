# Step 4

###  Creating a loyalty flag for existing customers using the transform() and loc() functions.


## Criteria used to create the flag are as following: 

### If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”

### If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”

### If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

## Step 4.1

```python
# 1st Create a 'max_order' column, into which will be placed the results of the aggragetion

# 2nd The ords_prods_merge dataframe is grouped by the “user_id” column

# 3rd the transform() function is applied on the “order_number” column with the np.max argument 

# After getting a 'FutureWarning' from Python, I removed the 'np' from the transform function, 
## instead transform(np.max), I've used the argument 'max' within the function transform('max')

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
# check the head

ords_prods_merge.head(15)
```

## Step 4.2

```python
# Create a 'loyalty_flag' column using the derived column from earlier step ('max_order')
# (a flag for customers with many orders.)

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# not to assign any options regarding the maximum number of rows to display

pd.options.display.max_rows = None
```

```python
# Check the counts

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# Step 6

### Create a spending flag for each user based on the average price across all their orders using the following criteria:

### 1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

### 2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# 1st Create a 'price_mean' column, into which will be placed the results of the aggregation

# 2nd The ords_prods_merge dataframe is grouped by the “user_id” column

# 3rd the transform() function is applied on the “prices” column with the 'max' argument 

ords_prods_merge['price_mean'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check the head

ords_prods_merge.head(15)
```

```python
# Create a 'spending_flag' column using the column 'price_mean'

ords_prods_merge.loc[ords_prods_merge['price_mean'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['price_mean']< 10, 'spending_flag'] = 'Low spender'
```

```python
# Check counts

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# check the head, but only with the columns of interet in order to checking the accuracy

ords_prods_merge[['user_id', 'price_mean', 'spending_flag']].head(100)
```

# Step 7 determine frequent versus non-frequent customers.

## Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_last_order” column.


### The criteria for the flag should be as follows:

### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# 1st Create a 'days_since_last_order_median' column, into which will be placed the results of the aggragetion

# 2nd Group the ords_prods_merge dataframe by the 'user_id' column 

# 3rd the transform() function is applied on the 'days_since_last_order' column with the 'median' argument 

ords_prods_merge['days_since_last_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
# Check the head with the new column

ords_prods_merge.head(10)
```

```python
# Create a 'frequency' column using the column 'days_since_last_order_median'

ords_prods_merge.loc[ords_prods_merge['days_since_last_order_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['days_since_last_order_median'] > 10) & (ords_prods_merge['days_since_last_order_median'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['days_since_last_order_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#check the counts

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# check the head, but only with the columns of interet in order to checking the accuracy

ords_prods_merge[['user_id', 'days_since_last_order_median', 'frequency_flag']].head(50)
```
